```python
class Printable:
    def __enter__(self):
        print('enter')

    def __exit__(self, type, value, traceback):
        print('exit')
```

```python
from contextlib import contextmanager


@contextmanager
def printable():
    print('enter')
    try:
        yield
    finally:
        print('exit')
```

```python
with Printable():
    print('abc')
```

```python
with printable():
    print('abc')
```

```python
try:
    # Попытка выполнить операцию, которая вызовет исключение
    1 / 0
except ZeroDivisionError as e:
    # Обрабатываем деление на ноль и поднимаем новое исключение
    raise ValueError("Неверные данные!") from e
```

```python
try:
    # Попытка выполнить операцию, которая вызовет исключение
    1 / 0
except ZeroDivisionError as e:
    # Обрабатываем деление на ноль и поднимаем новое исключение
    raise ValueError("Неверные данные!") from None
```

```python
try:
    # Попытка выполнить операцию, которая вызовет исключение
    1 / 0
except ZeroDivisionError as e:
    # Обрабатываем деление на ноль и поднимаем новое исключение
    raise ValueError("Неверные данные!")
```

```python
def divide(a, b):
    try:
        # Попытка выполнить деление
        result = a / b
    except ZeroDivisionError:
        # Обработка деления на ноль
        print("Ошибка: Деление на ноль!")
        return None        
    # Выполняется только если в блоке try не было исключений
    print("Деление выполнено успешно.")
    return result

# Пример 1: Корректное деление
print(divide(10, 2))  # Вывод: Деление выполнено успешно.

# Пример 2: Деление на ноль
print(divide(10, 0))  #
```

```python
def my_dec(f):
    def foo(*args, **kwargs):
        print('start')
        print(f(*args, **kwargs))
        print('end')
    return foo
```
